```python
# Contar el número de valores duplicados en la columna 'category'
total_duplicados = oscar_df.duplicated(subset=['category']).sum()
print("Número total de valores duplicados en la columna 'category':", total_duplicados)

# Encontrar los valores duplicados específicos en la columna 'category'
duplicados_unicos = oscar_df[oscar_df.duplicated(subset=['category'], keep=False)]['category'].value_counts()
print("Valores duplicados en la columna 'category':\n", duplicados_unicos)
```
