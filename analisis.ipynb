```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el CSV en un DataFrame
df_actores = pd.read_csv('data/processed/tabla_actores.csv')

df_actores['year_ceremony'].unique()

# Verificar el rango de años en el DataFrame
print(df_actores['year_ceremony'].min(), df_actores['year_ceremony'].max())

# Convertir la columna 'year_ceremony' a tipo numérico
df_actores['year_ceremony'] = pd.to_numeric(df_actores['year_ceremony'], errors='coerce')

# Crear una gráfica de dispersión de la edad de los actores por año de ceremonia, coloreada por raza
plt.figure(figsize=(14, 8))
sns.scatterplot(data=df_actores, x='year_ceremony', y='edad', hue='raza', style='winner', palette='Set1', s=100)
plt.title('Edad de los Actores por Año de Ceremonia')
plt.xlabel('Año de Ceremonia')
plt.ylabel('Edad')
plt.legend(title='Raza')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el CSV en un DataFrame
df_actores = pd.read_csv('data/processed/tabla_actores.csv')

# Convertir la columna 'edad' a tipo numérico
df_actores['edad'] = pd.to_numeric(df_actores['edad'], errors='coerce')

# Calcular la media de edad
mean_age = df_actores['edad'].mean()

# Crear un histograma con la curva de distribución de las edades de los actores
plt.figure(figsize=(14, 8))
sns.histplot(df_actores['edad'], kde=True, color='blue', bins=30)

# Añadir una línea vertical para la media de edad
plt.axvline(mean_age, color='red', linestyle='--', linewidth=2, label=f'Media de Edad: {mean_age:.2f}')

# Añadir títulos y etiquetas
plt.title('Distribución de las Edades de los Actores')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()
```
