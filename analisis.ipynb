```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el CSV en un DataFrame
df_actores = pd.read_csv('data/processed/tabla_actores.csv')

df_actores['year_ceremony'].unique()

# Verificar el rango de aos en el DataFrame
print(df_actores['year_ceremony'].min(), df_actores['year_ceremony'].max())

# Convertir la columna 'year_ceremony' a tipo numrico
df_actores['year_ceremony'] = pd.to_numeric(df_actores['year_ceremony'], errors='coerce')

# Crear una grfica de dispersin de la edad de los actores por ao de ceremonia, coloreada por raza
plt.figure(figsize=(14, 8))
sns.scatterplot(data=df_actores, x='year_ceremony', y='edad', hue='raza', style='winner', palette='Set1', s=100)
plt.title('Edad de los Actores por Ao de Ceremonia')
plt.xlabel('Ao de Ceremonia')
plt.ylabel('Edad')
plt.legend(title='Raza')
plt.grid(True)
plt.show()
```
